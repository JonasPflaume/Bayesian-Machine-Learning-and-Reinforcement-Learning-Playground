### show the mutinomial distribution and it's conjugate dirichlet distribution in P76

#### here we show if we only show one axis in mu, the probability will be like

```python
mu1_ = np.linspace(0,1,100)
mu2_ = 1.0 - mu1_
prob_=[]
for mu1,mu2 in zip(mu1_,mu2_):
    prob = muti_distribution([9,1],[mu1,mu2],10)
    prob_.append(prob)
plt.plot(mu1_, prob_, '-r')
```

#### direchlet distrubution below just like P78 left diagramm

#### what likelihood function looks like, here data were generated by mu=[0.3, 0.7], we could see likelihood value is maximum when mu1=0.3 corresponding mu2=0.7

```python
N = 20
mu = [0.3,0.7]
# produce a artificial dataset
Data = np.zeros((N, 2))
index = [np.random.choice([0,1], p=mu) for _ in range(N)]
for row, ind in enumerate(index):
    Data[row,ind] = 1
# test the likelihood function
muti_likelihood(Data, mu)

prob_like=[]
for mu1,mu2 in zip(mu1_,mu2_):
    prob_l = muti_likelihood(Data,[mu1,mu2])
    prob_like.append(prob_l)
plt.plot(mu1_, prob_like, '-r')
```
